### Testing with Pandas

```
solo = pd.read_csv("amostras_solo.csv", sep=';', index_col=False, header=None)
solo.describe()
```

```
## Remove mean
solo = solo.sub(solo.mean(axis=0), axis=1)

solo.describe()
```

```
# Covariance Matrix
R = np.cov(solo.values, rowvar=False)
R
```

### Eigenvalues and Eigenvectors

```
w, v = np.linalg.eig(R)

print(f"Autovalores: \n {w}")
print(f"autovetores: \n {v}")
```

### Ordenando

```
#from max to min

indexes = np.argsort(w)[::-1]
w_ordered = w[indexes]
#indexes = [np.where(w == a)[0][0] for a in w_ordered]
```

```
v_ordered = v[:,indexes]
print(f"Autovalores ordenados: \n{w_ordered}")
print(f"Autovetores ordenados: \n{v_ordered}")

```

### Using GHA

```
class GHA():
  
  def __init__(self, n=1, j=4, i=4, learning=0.0001):
    
    self.n = n
    self.n_outputs = j
    self.n_inputs = i
    self.learning_rate = learning
    self.w = np.random.randn(j, i)
    self.y = np.zeros(j)
   
  def fit(self, x):
    
    for it in range(self.n):
      #for out in range(self.n_outputs):
        #Sprint(np.dot(self.w[out,:], x))
      self.y = np.dot(self.w, x)
      dW = (np.outer(self.y,x) - np.dot(np.tril(np.outer(self.y,self.y)),self.w))*self.learning_rate
      self.w += dW
  def brief(self):
    
    return self.y, self.w
```

```
gha = GHA(j=4)

for _ in range(10000):
  for x in solo.values:
    gha.fit(x)

y, wgha = gha.brief()
print(y)
print(wgha)

## um dos autovetores pelo GHA eh diferente
```
